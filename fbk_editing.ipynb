```python
with open('OLD COURSE.fbk', 'r') as inp_file:
    lines=inp_file.readlines()

with open('OLD COURSE_new.fbk', 'w') as out_file:
    for line in lines:
        if line[:2] == 'F1':
            cols = line.strip().split()
            
            if float(cols[4]) != 0:
                cols[4] = str(float(cols[4])+0.023)

            line = ' '.join(cols) + '\n'
        
        out_file.write(line)
```

```python
list1 = np.arange(241, 270)
list1 = list1.astype('str')

list2 = np.arange(270,297)
list2 = list2.astype('str')

list2_ts_aux = np.arange(1,44)
list2_ts_aux=list2_ts_aux.astype('str')

for i in range(len(list2_ts_aux)):
    if i < 9:
        list2_ts_aux[i] = 'TS000'+str(i+1)
    else:
        list2_ts_aux[i] = 'TS00'+str(i+1)

list2=list2.tolist()
list2_ts_aux=list2_ts_aux.tolist()

list2 = list2 + list2_ts_aux
list2 += ['TS0086', 'TS0087', 'TS0500', 'TS0501', 'TS0502', 'TS0523']

list3 = np.arange(1130, 1152)
list3 = list3.astype('str')

list4 = np.arange(2000, 3117)
list4 = list4.astype('str')

final_list = []

for val in list1:
    final_list.append('"' + val + '"')

for val in list2:
    final_list.append('"' + val + '"')

for val in list3:
    final_list.append('"' + val + '"')

for val in list4:
    final_list.append('"' + val + '"')
```
